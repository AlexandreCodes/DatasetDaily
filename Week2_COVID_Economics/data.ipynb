```python
import pandas as pd
```

```python
# Load data. 
economic_data = pd.read_excel('data/economics.xlsx')  
economic_data.tail(15)
```

```python
# Perform EDA, visualize, etc...
import plotly.graph_objects as go


# Sample Figure. Unemployment. 
fig = go.Figure()

fig.add_trace(go.Scatter(x=economic_data['observation_date'], y=economic_data['unemployment_rate'],
                    mode='lines',
                    name='Unemployment Rate',showlegend=True))


fig.update_layout(
    xaxis=dict(
        showline=True,
        showgrid=False,
        showticklabels=True,
        linecolor='rgb(204, 204, 204)',
        linewidth=2,
        ticks='outside',
        tickfont=dict(
            family='Arial',
            size=12,
            color='rgb(82, 82, 82)',
        ),
    ),
    yaxis=dict(
        showgrid=False,
        zeroline=False,
        showline=False,
        showticklabels=True,
    ),
    autosize=False,
    margin=dict(
        l=100,
        r=20,
        t=110,
    ),
    plot_bgcolor='white'
)
fig.show()
```
