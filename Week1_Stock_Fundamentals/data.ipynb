```python
import requests
import pandas as pd
from IPython.display import display
pd.options.display.max_columns = None

api_key=''
```

```python
tickers = ['MSFT', 'AAPL', 'ORCL', 'IBM', 'GOOG', 'INTC', 'IBM', 'HPQ', 'DELL', 'AMZN']
```

```python
ticker = 'MSFT'
date = '2019-02-04'

close_price
years = ['2005','2006','2007','2008', '2009', '2010']
months_days = ['-01-22','-04-22','-07-22','-10-22']

import datetime

for x in years:
    for y in months_days: 
        date = x+y
        date = datetime.datetime.fromisoformat(date)

        if date.weekday() == 5:
            date = date + datetime.timedelta(days=2)
        if date.weekday() == 6: 
            date = date + datetime.timedelta(days=1)
        print(date)
        url = 'https://api.polygon.io/v1/open-close/' + ticker + "/" + str(date) + '?apiKey=' + api_key
        close_price = requests.get(url).json()['close']
        print(close_price)
```
